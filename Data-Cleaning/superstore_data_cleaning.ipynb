```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('sampleSuperstore.csv')

# Display basic information
print(df.head())
print(df.info())
```

```python
df.shape
```

```python
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
# Check for duplicate rows
print("Number of duplicate rows:", df.duplicated().sum())

# Check for duplicates in specific columns
print("\nDuplicates by Order ID (if exists):")
print(df.duplicated(subset=['City', 'State', 'Postal Code']).sum())

# Display duplicate rows if any exist
if df.duplicated().any():
    print("\nDuplicate rows:")
    print(df[df.duplicated(keep=False)].sort_values(by=list(df.columns)))
else:
    print("\nNo duplicate rows found in the dataset.")
```

```python
#Remove duplicates if necessary
df.drop_duplicates(inplace=True)
```

```python
print("After removing duplicates, number of duplicate rows:", df.duplicated().sum())
```

```python
print(df.dtypes)
```

```python

unique_counts = df.nunique()
print(unique_counts)
# print a header for columns that are unique across all rows
print("\nColumns with all unique values:")
print(unique_counts[unique_counts == len(df)])
print(unique_counts[unique_counts == len(df)])
```

```python
df['Segment'].unique()
```

```python
text_cols = df.select_dtypes(include='object').columns

for col in text_cols:
    print(col)
    print(df[col].unique())
    print()
```
